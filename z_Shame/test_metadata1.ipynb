```python
import requests
url = "http://api.us.socrata.com/api/catalog/v1?domains=www.chattadata.org"
resp = requests.get(url)
resp.raise_for_status()
resp
print("stop")
```

```python
data = resp.json()
res = data['results']
print("stop")
```

```python
c_list = []
for dicts in res:
    x = dicts['resource']
    i = 0
    while i < len(x):
        tup = (x['name'], x['columns_name'][i], x['columns_datatype'][i] )
        c_list.append(tup)
        i+=1
```

```python
c_list = [
    (dicts['resource']['name'], dicts['resource']['columns_name'][i], dicts['resource']['columns_datatype'][i])
    for dicts in res
    for i in range(len(dicts['resource']['columns_name']))
]
```

```python
for c in c_list:
    print(c)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(c_list, columns=["table_name", "column_name", "data_type"])
```
